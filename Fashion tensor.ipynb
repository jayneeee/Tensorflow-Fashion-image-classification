```python
import tensorflow as tf
from tensorflow import keras 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = keras.datasets.fashion_mnist
(train_images , train_labels), (test_images , test_labels ) = data.load_data()
class_names = ['T-shirts' , 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal,'
              ,'Shirt', 'Sneaker', 'Bag', 'Ankle Boot']
```

```python
#normalize/ scale the data 
train_images = train_images/255.0
test_images = test_images/255.0
#255 is the rgb scale 
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)), #28 by 28 is the pixel of the image
    # the data is in 2D-array form for each input hence need to be flatten out to be parsed into the model
    keras.layers.Dense(128, activation = 'relu'),#the hidden layer #128 is the number of neuron, a fully connected layer
    keras.layers.Dense(10 , activation = 'softmax') #the output layer
    ])

model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy' , metrics = ['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs = 5) #epochs how many times the model is going to see same the input 

test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Accuracy: ' ,test_acc)
```

```python
prediction = model.predict(test_images)
print(class_names[np.argmax(prediction[0])])
```

```python
for i in range(len(prediction)):
    print(class_names[np.argmax(prediction[i])])
```

```python
for i in range(5):
    plt.grid(False)
    plt.imshow(test_images[i], cmap = plt.cm.binary)
    plt.xlabel('Actual: ' + str(test_labels[i]))
    plt.title(class_names[np.argmax(prediction[i])])
    plt.show()
```
